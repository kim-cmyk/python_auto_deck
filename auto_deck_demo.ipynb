## Build all 6 themes (optional)
Generate one `.pptx` per theme from the same content so you can compare the colours.

```python
import os, re

for theme in THEMES:
    raw = open(INPUT_JSON, encoding='utf-8').read()
    # temporarily swap the THEME value, then build
    swapped = re.sub(r'("THEME"\s*:\s*")[^"]*(")', r'\g<1>' + theme + r'\g<2>', raw, count=1)
    tmp = f'_tmp_{theme}.json'
    open(tmp, 'w', encoding='utf-8').write(swapped)
    build_presentation(tmp, f'deck_{theme}.pptx')
    os.remove(tmp)
    print('Created deck_' + theme + '.pptx')
```
